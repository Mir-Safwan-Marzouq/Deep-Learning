```python
# Techniques to tackle overfitting and underfitting problem is called Regularization
# dropout regularization technique is used to tackle overfitting problem in deep learning field

# to avoid overfit I can randomly drop some neurons in hidden layers

# How dropout helps with overfitting?
## It can't rely on one input as it might be dropped out at random
## Neurons will not learn redundant details of inputs



```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import warnings
warnings.filterwarnings('ignore')

df = pd.read_csv('/content/sonar_dataset.csv', header=None) #has no column name so header=none
df.sample(5)
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
df.columns
```

```python
df[60].value_counts()
```

```python
X = df.drop(60, axis='columns')
y = df[60]
y.head()
```

```python
X
```

One Hot Encoding for y

```python
#y = pd.get_dummies(y)
#y.sample(5)
```

```python
y = pd.get_dummies(y, drop_first=True)
y.sample(5)
```

```python
y.value_counts()
```

Divide the data into train and test

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=1)
```

```python
X_train.shape, X_test.shape
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
model = keras.Sequential([
    #how many neurons in input layer=60, because there are 60 columns in total
    keras.layers.Dense(60, input_dim=60, activation='relu'),
    keras.layers.Dense(30, activation='relu'),
    keras.layers.Dense(15, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')
])

model.compile(loss= 'binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])

model.fit(X_train, y_train, epochs=100, batch_size=8) #in each iteration I feed 8 samples, calculate error and 
# then back propagate
```

```python
model.evaluate(X_test, y_test)
```

```python
y_pred = model.predict(X_test).reshape(-1)
print(y_pred[:10])

# round the values to nearest integer ie 0 or 1
y_pred = np.round(y_pred)
print(y_pred[:10])
```

```python
y_test[:10]
```

```python
from sklearn.metrics import confusion_matrix , classification_report

print(classification_report(y_test, y_pred))
```

```python
modeld = keras.Sequential([
    #how many neurons in input layer=60, because there are 60 columns in total
    keras.layers.Dense(60, input_dim=60, activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(30, activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(15, activation='relu'),
    keras.layers.Dropout(0.2),
    keras.layers.Dense(1, activation='sigmoid')
])

modeld.compile(loss= 'binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])

modeld.fit(X_train, y_train, epochs=100, batch_size=8) 
```

```python
modeld.evaluate(X_test, y_test)
```

```python
y_pred = modeld.predict(X_test).reshape(-1)
y_pred = np.round(y_pred)
from sklearn.metrics import confusion_matrix , classification_report

print(classification_report(y_test, y_pred))
```
